```python
# Import Libraries
import pandas as pd 
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.statsmodels.org/stable/index.html'

def getPage(url):
    request = requests.get(url).text
    soup = BeautifulSoup(request,'html.parser').prettify()
    return soup

print(getPage(url)
)



```
