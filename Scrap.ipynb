```python
# Data implementation + cleaning 

stock = "AMD"

class datascrapper():
    
    def valuation_measures(stock):
        collected_data = []
        stats = webRequests.getStatistics(stock)
        
        for elem in stats.find_all('div','Fl(start) smartphone_W(100%) W(50%)'):
            a = elem.findAll('td','Fw(500) Ta(end) Pstart(10px) Miw(60px)')
            for el in a:
                collected_data.append(el.text)

        vlist = []

        for el in valuation_measures:
            vlist.append(el)
            
        i=0
        while i < len(collected_data):
            valuation_measures[vlist[i]] = collected_data[i]
            i = i+1
        
        return valuation_measures
    
    
    
    
    
        
df = pd.DataFrame(datascrapper.valuation_measures("AMD"),index=[0])
df.head()
```
